# 3D Görselleştirme Türleri ve Kullanım Alanları
3D görselleştirme teknikleri, verilerin daha kapsamlı ve anlaşılır bir şekilde sunulmasını sağlar. Bu teknikler, özellikle karmaşık ve çok boyutlu veri kümelerinin analizinde ve sunumunda büyük avantajlar sunar. İşte yaygın olarak kullanılan bazı 3D görselleştirme türleri ve kullanım alanları:

1. 3D Çubuk Grafik (3D Bar Plot)
Neden Kullanılır:

- Kategorik verileri ve ilişkilerini üç boyutta görselleştirmek için idealdir.
- Farklı kategoriler arasındaki farkları ve trendleri net bir şekilde gösterebilir.

Kullanım Alanları:

- Satış verileri, üretim miktarları gibi iş analizlerinde.
- Sosyal bilimlerde anket sonuçlarının görselleştirilmesi.
- Finansal verilerin analizinde.

2. 3D Yüzey Grafiği (3D Surface Plot)

Neden Kullanılır:

- Sürekli veri kümelerinin üç boyutta eğilimlerini ve ilişkilerini görselleştirmek için kullanılır.
- Karmaşık fonksiyonların ve yüzeylerin görselleştirilmesini sağlar.

Kullanım Alanları:

- Fiziksel ve matematiksel modellemelerde.
- Jeolojik ve coğrafi analizlerde.
- Ekonomik modellerde değişkenler arasındaki ilişkilerin incelenmesi.

3. 3D Kontur Grafiği (3D Contour Plot)

Neden Kullanılır:

- Bir yüzeyin farklı yüksekliklerdeki kesitlerini görselleştirmek için kullanılır.
- Yüksekliğe bağlı olarak değişen değerlerin dağılımını gösterir.

Kullanım Alanları:

- Topografik haritalar ve coğrafi analizler.
- Isı haritaları ve termal analizlerde.
- Akışkanlar dinamiği ve hava durumu modellemeleri.

4. 3D Tel Çerçeve Grafiği (3D Wireframe Plot)

Neden Kullanılır:

- Bir yüzeyin tel çerçeve modeli ile basit ve anlaşılır bir şekilde görselleştirilmesini sağlar.
- Veri noktalarının düzenini ve yapısını net bir şekilde gösterir.

Kullanım Alanları:

- Mühendislik ve mimari tasarımlar.
- Bilgisayar grafikleri ve oyun geliştirme.
- Bilimsel araştırmalar ve simülasyonlar.

5. 3D Üçgen Yüzey Grafiği (3D Tri-Surface Plot)

Neden Kullanılır:

- Rastgele dağıtılmış veri noktalarının üç boyutta bir yüzey oluşturacak şekilde görselleştirilmesini sağlar.
- Veri noktaları arasındaki ilişkilerin ve eğilimlerin anlaşılmasını kolaylaştırır.

Kullanım Alanları:

- Jeoloji ve jeofizik analizler.
- İklim bilimleri ve çevresel modellemeler.
- Makine öğrenimi ve veri bilimi projeleri.

# 1- 3d veri görselleştirme ortamını oluşturma

```python
# Gerekli kütüpanlerin import edilmesi

from mpl_toolkits import mplot3d
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Görselleştirmenin yapılacağı 3d ortam

fig = plt.figure()
ax = plt.axes(projection='3d')
```

# 2 - Oluşturduğumuz veri ortamında nokta değereler veripi burdan veri ortalamasından çizgi grafiği geçirmek.

```python
ax = plt.axes(projection='3d')

# Data for a three-dimensional line
zline = np.linspace(0, 15, 1000)
xline = np.sin(zline)
yline = np.cos(zline)
ax.plot3D(xline, yline, zline, 'gray')

# Data for three-dimensional scattered points
zdata = 15 * np.random.random(100)
xdata = np.sin(zdata) + 0.1 * np.random.randn(100)
ydata = np.cos(zdata) + 0.1 * np.random.randn(100)
ax.scatter3D(xdata, ydata, zdata, c=zdata, cmap='Greens');
```

# 3 - Karmaşık veri seti üzerinden 3d veri seti ile 2d sütün grafiklerinin yetmediği durumlarda kullanılabilir.

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Data for a three-dimensional bar plot
_x = np.arange(4)
_y = np.arange(3)
_xx, _yy = np.meshgrid(_x, _y)
x, y = _xx.ravel(), _yy.ravel()

top = x + y
bottom = np.zeros_like(top)
width = depth = 1

ax.bar3d(x, y, bottom, width, depth, top, shade=True)
ax.set_title('3D Bar Plot')
plt.show()
```

# 4 - Bu garafik harita özelikleri yada verinin artış ve azalışdurumlarını 3d artış ve azalışını görselleştirmede işinize yarayabilir.

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Data for a three-dimensional surface
X = np.linspace(-5, 5, 100)
Y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(X, Y)
Z = np.sin(np.sqrt(X**2 + Y**2))

ax.plot_surface(X, Y, Z, cmap='viridis')
ax.set_title('3D Surface Plot')
plt.show()
```

```python
# 5 - Bu alanda verinin 3d uzayda grafik gösterilemsi. 1.0 durumda anlayamadığınız görseli 1.1 de tan anlaşılır.
```

```python
# 1.0 görselleştirme.

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Data for a three-dimensional triangular surface plot
X = np.random.rand(100)
Y = np.random.rand(100)
Z = np.sin(X) + np.cos(Y)

ax.plot_trisurf(X, Y, Z, cmap='jet')
ax.set_title('3D Tri-Surface Plot')
plt.show()
```
